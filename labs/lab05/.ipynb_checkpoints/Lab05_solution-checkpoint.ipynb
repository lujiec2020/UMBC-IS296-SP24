## Write code to create plots so that each of the question can be answered, you do not need to give the actual answer

```python
#look up table with columns Month (Jan, Feb ...) and Month ID (1,2,3...)
month_lookup = Table.read_table('./month_lookup.csv')
#raw weather data
weather_raw = Table.read_table('./weather.csv')
#join raw weather data and month look up table
weather=weather_raw.join('Month',month_lookup,'Month').select('Month','MonthID',
                                                              'Day Index','Quarter ID','High','Low','Rain')
weather.show(5)
```

### Q1. Are there more raindy than  non raindy days?

```python
weather.group('Rain').barh('Rain')
```

### Q2. What percentage of days  have a high above 72 degree? Hint: use bins parameter


```python
weather.hist('High',bins=np.arange(30,100,2))
```

### Q3. Do hotter days tend to  also have hotter nights?

```python
weather.scatter('High','Low')
```

```python
weather.select('Month','High').group('Month').show()
```

```python
weather.select('MonthID','Month').group(('MonthID','Month')).show()
```

```python
weather.select('MonthID','Month','High','Low').group(('MonthID','Month'),np.mean).show()
```

```python
Q1 = weather.where('Quarter ID', 1)
Q1.select('Month','High').group('Month',np.mean).show()
```

```python
Q1 = weather.where('Quarter ID',1)
Q1.select('Month','Rain').group(('Month','Rain')).show()
```
